```python
import pandas as pd

data = pd.read_csv(r'.\Data\Dataset.csv')
data.head()
data.shape
```

```python
#Scatter Plot
data.plot.scatter(x='radius_mean', y='texture_mean', title='Dataset')
```

```python
#Line Chart
data.drop(['radius_mean'], axis=1).plot.line(title='Dataset')
```

```python
#Histogram
data['radius_mean'].plot.hist()
```

```python
#Bar chart
data['radius_mean'].value_counts().sort_index().plot.bar()
```

```python
#Horizontal Bar chart
data['radius_mean'].value_counts().sort_index().plot.barh()
```
