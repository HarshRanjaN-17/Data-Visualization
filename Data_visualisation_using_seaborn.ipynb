```python
import pandas as pd
import seaborn as sns

data = pd.read_csv('.\Data\Dataset.csv')
data.head()
```

```python
# y includes our labels and x includes our features
y = data.diagnosis                          # M or B 
list = ['Unnamed: 32','id','diagnosis']
x = data.drop(list,axis = 1 )
x.head()
```

```python
import matplotlib.pyplot as plt
```

```python
ax = sns.countplot(y,label="Count")       # M = 212, B = 357
B, M = y.value_counts()
print('Number of Benign: ',B)
print('Number of Malignant : ',M)
```

```python
#scatter plot
sns.scatterplot(x='radius_mean', y='texture_mean', data= x)
```

```python
#highlightig the points by "variety"
sns.scatterplot(x='radius_mean', y='texture_mean', hue='diagnosis', data=data)
```

```python
#histogram
sns.distplot(data['radius_mean'], bins=10, kde=False)
```

```python
sns.distplot(data['radius_mean'], bins=10, kde=True)
```

```python
#bar chart
sns.countplot(data['radius_mean'])
```

```python
sns.set(style="white")
df = x.loc[:,['radius_worst','perimeter_worst','area_worst']]
g = sns.PairGrid(df, diag_sharey=False)
g.map_lower(sns.kdeplot, cmap="Blues_d")
g.map_upper(plt.scatter)
g.map_diag(sns.kdeplot, lw=3)
```

```python
#correlation map
f,ax = plt.subplots(figsize=(14, 14))
sns.heatmap(x.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
```
